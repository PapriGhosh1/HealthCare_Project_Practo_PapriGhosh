```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
HealthCare_Data = pd.read_csv("heart_disease_health_indicators_BRFSS2015_PapriGhosh.csv")
HealthCare_Data.head(10)
```

```python
Counts = HealthCare_Data["Smoker"].value_counts()
plt.pie(Counts, labels = Counts.index, autopct = "%1.1f%%")
plt.title("Distribution_SmokingStatus")
plt.show()
```

```python
Counts = HealthCare_Data["PhysActivity"].value_counts()
plt.pie(Counts, labels = Counts.index, autopct = "%1.1f%%")
plt.title("Distribution_PhysicalActivity")
plt.show()
```

```python
Counts = HealthCare_Data["Fruits"].value_counts()
plt.pie(Counts, labels = Counts.index, autopct = "%1.1f%%")
plt.title("Distribution_FruitConsumption")
plt.show()
```

```python
Age_Group = HealthCare_Data.groupby("Age")["HeartDiseaseorAttack"].mean()
plt.bar(Age_Group.index, Age_Group.values, color= "Skyblue")
plt.title("Heart Disease Prevalance By Age")
plt.xlabel("Age")
plt.ylabel("HeartDiseaseorAttack")
plt.show()
```

```python
Smoker_Group = HealthCare_Data.groupby("Smoker")["HeartDiseaseorAttack"].mean()
plt.bar(Smoker_Group.index, Smoker_Group.values, color= "Skyblue")
plt.title("Heart Disease Prevalance By Smoking")
plt.xlabel("Smoker")
plt.ylabel("HeartDiseaseorAttack")
plt.show()
```

```python
PhysicalActivity_Group = HealthCare_Data.groupby("PhysActivity")["HeartDiseaseorAttack"].mean()
plt.bar(PhysicalActivity_Group.index, PhysicalActivity_Group.values, color= "Skyblue")
plt.title("Heart Disease Prevalance By PhysicalActivity")
plt.xlabel("PhysActivity")
plt.ylabel("HeartDiseaseorAttack")
plt.show()
```

```python
Education_Group = HealthCare_Data.groupby("Education")["HeartDiseaseorAttack"].mean()
plt.bar(Education_Group.index, Education_Group.values, color= "Skyblue")
plt.title("Heart Disease Prevalance By Education")
plt.xlabel("Education")
plt.ylabel("HeartDiseaseorAttack")
plt.show()
```
