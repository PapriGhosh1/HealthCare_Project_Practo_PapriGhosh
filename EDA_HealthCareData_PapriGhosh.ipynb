```python
Age_Group = HealthCare_Data.groupby("Age")["HeartDiseaseorAttack"].mean()
plt.bar(Age_Group.index, Age_Group.values, color= "Skyblue")
plt.title("Heart Disease Prevalance By Age")
plt.xlabel("Age")
plt.ylabel("HeartDiseaseorAttack")
plt.show()
```

```python
Gender_Group = HealthCare_Data.groupby("Sex")["HeartDiseaseorAttack"].mean()
plt.bar(Gender_Group.index, Gender_Group.values, color = ["Skyblue", "Powderblue"], edgecolor = "Pink")
plt.title("Heart Disease Prevalance By Gender")
plt.xlabel("Gender")
plt.ylabel("HeartDiseaseorAttack")
plt.show()
```

```python
Education_Group = HealthCare_Data.groupby("Education")["HeartDiseaseorAttack"].mean()
plt.bar(Education_Group.index, Education_Group.values, color= "Skyblue")
plt.title("Heart Disease Prevalance By Education")
plt.xlabel("Education")
plt.ylabel("HeartDiseaseorAttack")
plt.show()
```
